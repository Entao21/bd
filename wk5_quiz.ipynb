## Question 1

Create the following tables using the corresponding shapefiles.
- nyc_neighborhoods
- nyc_census_blocks
- nyc_homicides
- nyc_streets
- nyc_subway_stations

```python
import duckdb
from pathlib import Path

# Magic sql kept giving error so switching to conda syntax

DATA_DIR = Path("/content/data")
DATA_DIR.mkdir(parents=True, exist_ok=True)

shps = list(DATA_DIR.rglob("*.shp"))
print(f"Found {len(shps)} shapefiles under {DATA_DIR}")
for p in shps[:15]:
    print(" -", p)

def find_shp(keyword: str) -> Path:
    key = keyword.lower()
    matches = [p for p in shps if key in p.name.lower()]
    if not matches:
        raise FileNotFoundError(f"Cannot find shapefile with keyword '{keyword}'. "
                                f"Available names include: {[p.name for p in shps[:20]]}")
    return sorted(matches, key=lambda x: len(str(x)))[0]

paths = {
    "nyc_neighborhoods": find_shp("neighborhood"),
    "nyc_census_blocks": find_shp("census_blocks"),
    "nyc_homicides": find_shp("homicide"),
    "nyc_streets": find_shp("streets"),
    "nyc_subway_stations": find_shp("subway_stations"),
}

print("\nChosen shapefiles:")
for k, v in paths.items():
    print(f"{k}: {v}")

con = duckdb.connect(database=":memory:")
con.execute("INSTALL spatial;")
con.execute("LOAD spatial;")

for table, shp_path in paths.items():
    con.execute(f"""
        CREATE OR REPLACE TABLE {table} AS
        SELECT * FROM ST_Read('{shp_path.as_posix()}');
    """)

counts = con.execute("""
    SELECT 'nyc_neighborhoods' AS table, COUNT(*) AS n FROM nyc_neighborhoods
    UNION ALL SELECT 'nyc_census_blocks', COUNT(*) FROM nyc_census_blocks
    UNION ALL SELECT 'nyc_homicides', COUNT(*) FROM nyc_homicides
    UNION ALL SELECT 'nyc_streets', COUNT(*) FROM nyc_streets
    UNION ALL SELECT 'nyc_subway_stations', COUNT(*) FROM nyc_subway_stations
""").df()
counts
```

```python
BOROUGH_COL = "BORONAME"
DATE_COL    = "INCIDENT_D"

q3 = f"""
SELECT
  {BOROUGH_COL} AS borough,
  EXTRACT(YEAR FROM CAST({DATE_COL} AS DATE)) AS year,
  COUNT(*) AS n_victims
FROM nyc_homicides
GROUP BY 1,2
ORDER BY n_victims DESC
LIMIT 10;
"""
con.execute(q3).df()
```

## Question 4:

Which subway station has the largest number of shooting victims within a 300 meter radius?


```python
con.execute("DESCRIBE nyc_subway_stations;").df()
```

```python
con.execute("SELECT * FROM nyc_subway_stations;").df()
```

```python
q4 = """
WITH gun_victims AS (
  SELECT
    geom,
    TRY_CAST(NUM_VICTIM AS INTEGER) AS num_victim_int
  FROM nyc_homicides
  WHERE LOWER(WEAPON) = 'gun'
),
station_totals AS (
  SELECT
    s.NAME AS station_name,
    SUM(COALESCE(h.num_victim_int, 0)) AS shooting_victims
  FROM nyc_subway_stations s
  JOIN gun_victims h
    ON ST_DWithin(s.geom, h.geom, 300)
  GROUP BY 1
)
SELECT station_name
FROM station_totals
ORDER BY shooting_victims DESC
LIMIT 1;
"""
con.execute(q4).df()
```

```python
q4_top10 = """
WITH gun_victims AS (
  SELECT
    geom,
    TRY_CAST(NUM_VICTIM AS INTEGER) AS num_victim_int
  FROM nyc_homicides
  WHERE LOWER(WEAPON) = 'gun'
),
station_totals AS (
  SELECT
    s.NAME AS station_name,
    SUM(COALESCE(h.num_victim_int, 0)) AS shooting_victims
  FROM nyc_subway_stations s
  JOIN gun_victims h
    ON ST_DWithin(s.geom, h.geom, 300)
  GROUP BY 1
)
SELECT station_name, shooting_victims
FROM station_totals
ORDER BY shooting_victims DESC;
"""
con.execute(q4_top10).df()
```

## Question 5

what is the most densely populated residential street in new york? (total population of intersecting census blocks divided by the total length of the street)

```python
con.execute("DESCRIBE nyc_streets;").df()
```

```python
con.execute("""
SELECT *
FROM nyc_streets;
""").df()
```

```python
con.execute("SELECT * FROM nyc_census_blocks;").df()
```

```python
q5 = """
WITH streets_res AS (
  SELECT
    ID AS street_id,
    NAME AS street_name,
    geom,
    ST_Length(geom) AS street_len
  FROM nyc_streets
  WHERE TYPE = 'residential'
    AND NAME IS NOT NULL
),
street_block AS (
  SELECT DISTINCT
    s.street_id,
    b.BLKID,
    TRY_CAST(b.POPN_TOTAL AS DOUBLE) AS pop
  FROM streets_res s
  JOIN nyc_census_blocks b
    ON ST_Intersects(s.geom, b.geom)
),
pop_by_street AS (
  SELECT
    street_id,
    SUM(COALESCE(pop, 0)) AS pop_sum
  FROM street_block
  GROUP BY 1
)
SELECT s.street_name
FROM streets_res s
JOIN pop_by_street p
  ON s.street_id = p.street_id
ORDER BY (p.pop_sum / NULLIF(s.street_len, 0)) DESC
LIMIT 1;
"""
con.sql(q5)
```

```python
q5_check = """
WITH streets_res AS (
  SELECT
    ID AS street_id,
    NAME AS street_name,
    geom,
    ST_Length(geom) AS street_len
  FROM nyc_streets
  WHERE TYPE = 'residential'
    AND NAME IS NOT NULL
),
street_block AS (
  SELECT DISTINCT
    s.street_id,
    b.BLKID,
    TRY_CAST(b.POPN_TOTAL AS DOUBLE) AS pop
  FROM streets_res s
  JOIN nyc_census_blocks b
    ON ST_Intersects(s.geom, b.geom)
),
pop_by_street AS (
  SELECT
    street_id,
    SUM(COALESCE(pop, 0)) AS pop_sum
  FROM street_block
  GROUP BY 1
)
SELECT
  s.street_name,
  p.pop_sum,
  s.street_len,
  (p.pop_sum / NULLIF(s.street_len, 0)) AS density
FROM streets_res s
JOIN pop_by_street p
  ON s.street_id = p.street_id
ORDER BY density DESC
LIMIT 10;
"""
con.execute(q5_check).df()
```

This section uses internet speed data from [Ookla](https://github.com/teamookla/ookla-open-data); take a minute to read the documentation of the dataset.

Create a table called ookla_nyc which reads the data from this url:

https://storage.googleapis.com/qm2/CASA0025/ookla_nyc.parquet




```python

con.execute("INSTALL httpfs;")
con.execute("LOAD httpfs;")

con.execute("""
CREATE OR REPLACE TABLE ookla_nyc AS
SELECT *
FROM read_parquet('https://storage.googleapis.com/qm2/CASA0025/ookla_nyc.parquet');
""")
con.execute("SELECT * FROM ookla_nyc;").df()
```

```python
con.execute("""
CREATE OR REPLACE VIEW ookla AS
SELECT
  *,
  ST_GeomFromText(geometry) AS geom
FROM ookla_nyc;
""")
```

## Question 7

Which neighbourhood has the highest average download speed? use 'within' rather than intersects, and ignore tiles with fewer than 3 tests.

```python
q7 = """
SELECT
  n.NAME AS neighborhood,
  AVG(o.avg_d_kbps) AS avg_download_kbps
FROM nyc_neighborhoods n
JOIN ookla o
  ON ST_Within(o.geom, n.geom)
WHERE o.tests >= 3
GROUP BY 1
ORDER BY avg_download_kbps DESC
LIMIT 1;
"""
con.sql(q7)
```

## Question 8



What is the name of the subway station closest to the tile with the highest download speed?

```python
q8 = """
WITH max_speed AS (
  SELECT MAX(avg_d_kbps) AS mx
  FROM ookla
),
best_tiles AS (
  SELECT quadkey, geom, avg_d_kbps
  FROM ookla
  WHERE avg_d_kbps = (SELECT mx FROM max_speed)
),
nearest_station AS (
  SELECT
    s.NAME AS station_name,
    MIN(ST_Distance(s.geom, ST_Centroid(t.geom))) AS dist_to_best_tile
  FROM nyc_subway_stations s
  CROSS JOIN best_tiles t
  GROUP BY 1
)
SELECT station_name
FROM nearest_station
ORDER BY dist_to_best_tile ASC
LIMIT 1;
"""
con.execute(q8).df()
```

## Question 9

Which neighbourhood has the highest number of devices per capita? Use only spatial intersections.

```python
q9 = """
WITH devices_by_n AS (
  SELECT
    n.NAME AS neighborhood,
    SUM(COALESCE(o.devices, 0)) AS devices_sum
  FROM nyc_neighborhoods n
  JOIN (
    SELECT DISTINCT quadkey, devices, geom
    FROM ookla
  ) o
    ON ST_Intersects(o.geom, n.geom)
  GROUP BY 1
),
pop_by_n AS (
  SELECT
    n.NAME AS neighborhood,
    SUM(COALESCE(TRY_CAST(b.POPN_TOTAL AS DOUBLE), 0)) AS pop_sum
  FROM nyc_neighborhoods n
  JOIN (
    SELECT DISTINCT BLKID, POPN_TOTAL, geom
    FROM nyc_census_blocks
  ) b
    ON ST_Intersects(b.geom, n.geom)
  GROUP BY 1
)
SELECT d.neighborhood
FROM devices_by_n d
JOIN pop_by_n p USING (neighborhood)
ORDER BY (d.devices_sum / NULLIF(p.pop_sum, 0)) DESC
LIMIT 1;
"""
con.execute(q9).df()
```

```python
con.execute("""
WITH pop_by_n AS (
  SELECT
    n.NAME AS neighborhood,
    SUM(COALESCE(TRY_CAST(b.POPN_TOTAL AS DOUBLE), 0)) AS pop_sum
  FROM nyc_neighborhoods n
  JOIN nyc_census_blocks b
    ON ST_Intersects(b.geom, n.geom)
  GROUP BY 1
)
SELECT *
FROM pop_by_n
WHERE pop_sum = 0
ORDER BY neighborhood;
""").df()
```

```python
q9_top10 = """
WITH devices_by_n AS (
  SELECT
    n.NAME AS neighborhood,
    SUM(COALESCE(o.devices, 0)) AS devices_sum
  FROM nyc_neighborhoods n
  JOIN (SELECT DISTINCT quadkey, devices, geom FROM ookla) o
    ON ST_Intersects(o.geom, n.geom)
  GROUP BY 1
),
pop_by_n AS (
  SELECT
    n.NAME AS neighborhood,
    SUM(COALESCE(TRY_CAST(b.POPN_TOTAL AS DOUBLE), 0)) AS pop_sum
  FROM nyc_neighborhoods n
  JOIN (SELECT DISTINCT BLKID, POPN_TOTAL, geom FROM nyc_census_blocks) b
    ON ST_Intersects(b.geom, n.geom)
  GROUP BY 1
)
SELECT
  d.neighborhood,
  d.devices_sum,
  p.pop_sum,
  (d.devices_sum / NULLIF(p.pop_sum, 0)) AS devices_per_capita
FROM devices_by_n d
JOIN pop_by_n p USING (neighborhood)
WHERE p.pop_sum>0
ORDER BY devices_per_capita DESC;
"""
con.execute(q9_top10).df()
```

## Question 10


What is the average upload speed for tiles that are within 200 meters of L train stops and intersect with census blocks where the black population is above 50%?

```python
q10 = """
WITH l_stops AS (
  SELECT geom
  FROM nyc_subway_stations
  WHERE strpos(ROUTES, 'L') > 0
),
majority_black_blocks AS (
  SELECT geom
  FROM nyc_census_blocks
  WHERE (TRY_CAST(POPN_BLACK AS DOUBLE) / NULLIF(TRY_CAST(POPN_TOTAL AS DOUBLE), 0)) > 0.5
),
eligible_tiles AS (
  SELECT DISTINCT o.quadkey, o.avg_u_kbps
  FROM ookla o
  JOIN l_stops s
    ON ST_DWithin(o.geom, s.geom, 200)
  JOIN majority_black_blocks b
    ON ST_Intersects(o.geom, b.geom)
)
SELECT AVG(avg_u_kbps) AS avg_upload_kbps
FROM eligible_tiles;
"""
con.execute(q10).df()
```
